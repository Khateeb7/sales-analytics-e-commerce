```python
ax = sns.countplot(x='Gender',data=df)
```

```python
ax = sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x= 'Gender',y= 'Amount',data= sales_gen)
```

```python
#FROM ABOVE GRAPHS WE CAN SEE THAT MOST BUYERS ARE FEMALES AND EVEN PURCHASING POWER OF FEMALES IS GREATER THAN MALES
```

```python
ax= sns.countplot(x='Age Group', data=df, hue='Gender')
```

```python
ax= sns.countplot(x='Age Group', data=df, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

```python
#WE CAN CONCLUDE THAT BOTH MALES AND FEMALES FROM 26-35 GROUP SHOP THE MOST.AGE GROUP 26-35 ALSO SPENDS THE MOST AMONG ALL AGE GROUPS.
```

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

```python
#From above 2 graphs we can see that most orders as well as most amount spent is by customers from states Uttar Pradesh,Maharashtra and Karnataka.
```

```python
ax=sns.countplot(x='Marital_Status',data=df)

sns.set(rc={'figure.figsize':(6,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_maritalstatus=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_maritalstatus,x='Marital_Status',y='Amount',hue='Gender')
```

```python
#WE CAN CONCLUDE THAT THERE ARE MORE ORDERS AND AMOUNT SPENT FROM MARRIED PPL THAN NON MARRIED PPL.AND IN MARRIED PPL,MARRIED FEMALES HAVE SPENT QUITE MORE THAN MARRIED MALES
```

```python
ax=sns.countplot(x='Occupation',data=df)

sns.set(rc={'figure.figsize':(25,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_occupation,x='Occupation',y='Amount')
```

```python
#WE CAN CONCLUDE THAT PEOPLE FROM IT SECTOR,HEALTHCARE AND AVIATION HAVE MOST PURCHASING POWER AMONG ALL OCCUPATIONS
```

```python
ax=sns.countplot(x='Product_Category',data=df)

sns.set(rc={'figure.figsize':(20,6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_productcategory=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_productcategory,x='Product_Category',y='Amount')
```

```python
#WE CAN DRAW CONCLUSION THAT CLOTHING & APPAREL,FOOD AND ELECTRONICS & GADGETS GOT MOST ORDERS WHILE FOOD,CLOTHING & APPAREL AND ELECTRONICS & GADGETS GOT MOST AMOUNT IN ALL PRODUCT CATEGORIES. 
```

```python
sales_productid=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_productid,x='Product_ID',y='Orders')
```
